```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import csv to df
salary_df1 = pd.read_csv("../exported_data/salary_data1.csv")
salary_df1.head()
```

```python
del salary_df1["Unnamed: 0"]
salary_df1
```

```python
# sets X as every columns except koi_disposition
X = salary_df1.drop("SalaryUSD", axis=1)
# sets y to koi_disposition
y = salary_df1["SalaryUSD"].values.reshape(-1, 1)

print(X.shape, y.shape)
```

```python
X.keys()
```

```python
# X_encoded = pd.get_dummies(X, columns=[["country", "primarydatabase", "employmentstatus", "jobtitle", "managestaff", "otherpeopleonyourteam", "education", "educationiscomputerrelated", "certifications", "telecommutedaysperweek", "employmentsector", "region"]])
X_encoded = pd.get_dummies(X)
X_encoded
```

```python
y = y.astype(int)
y.dtype
```

```python
for col in X_encoded.keys():
    X_encoded[col] = X_encoded[col].astype(int)
X_encoded.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
# fits random forest model to X_encoded and y
rf = rf.fit(X_encoded, y)
# returns the importance of each column to predicting the outcomes
importances = rf.feature_importances_
importances
```

```python
# sorts each column in order of importance
sorted(zip(importances, X.keys()), reverse=True)
```

```python
# Choose the most importance features- this will be the new "X" (selected_features)
selected_features = salary_df1[['YearsWithThisDatabase', 'Country', 'PrimaryDatabase', 'EmploymentStatus', 'Region', 'YearsWithThisTypeOfJob', 'EducationIsComputerRelated', 'OtherPeopleOnYourTeam', 'TelecommuteDaysPerWeek', 'EmploymentSector', 'ManageStaff', 'Education']]
```

```python
X_encoded2 = pd.get_dummies(selected_features)
X_encoded2
```

```python
for col in X_encoded2.keys():
    X_encoded2[col] = X_encoded2[col].astype(int)
X_encoded2.dtypes
```

```python
# split the data up between training and testing data (3/4:1/4)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded2, y, random_state=42)
```

```python
# Scales the X data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
import numpy as np
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=16, activation='relu', input_dim=144))
model.add(Dense(units=144, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```
