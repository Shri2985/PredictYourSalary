```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import csv to df
salary_df2 = pd.read_csv("../exported_data/salary_data2_new.csv")

del salary_df2["Unnamed: 0"]
salary_df2 = salary_df2.dropna()
salary_df2 = salary_df2.drop(salary_df2[salary_df2.salary < 5000].index)
salary_df2.head()
```

```python
salary_df2["salary"].value_counts()
```

```python
# sets X as every columns except koi_disposition
X = salary_df2.drop("salary", axis=1)

# sets y to koi_disposition
y = salary_df2["salary"].astype(str).values.reshape(-1, 1)

print(X.shape, y.shape)
```

```python
X_encoded = pd.get_dummies(X)
print(X_encoded.shape, y.shape)
X_encoded
```

```python
y = y.astype(int)
y.dtype
```

```python
for col in X_encoded.keys():
    X_encoded[col] = X_encoded[col].astype(int)
X_encoded.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
# fits random forest model to X_encoded and y
rf = rf.fit(X_encoded, y)
# returns the importance of each column to predicting the outcomes
importances = rf.feature_importances_
importances
```

```python
# sorts each column in order of importance
sorted(zip(importances, X.keys()), reverse=True)
```
